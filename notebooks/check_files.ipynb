```python
import pickle
from cot_probing.typing import *
from transformers import AutoTokenizer
from pathlib import Path

dir = Path("../results/google--gemma-2-2b/snarks/bias-A_seed-0_total-10")

with open(dir / "tokenized_questions.pkl", "rb") as f:
    tokenized_qs = pickle.load(f)
with open(dir / "biased_context" / "tokenized_responses.pkl", "rb") as f:
    biased_tokenized_responses = pickle.load(f)
with open(dir / "unbiased_context" / "tokenized_responses.pkl", "rb") as f:
    unbiased_tokenized_responses = pickle.load(f)

tokenizer = AutoTokenizer.from_pretrained("google/gemma-2-2b")
```

```python
for i in range(len(tokenized_qs)):
    print(f"[Question {i+1}]")
    print()
    print(f"Question:\n`{tokenizer.decode(tokenized_qs[i].tokenized_question)}`")
    print(f"Biased:\n`{tokenizer.decode(biased_tokenized_responses[i])}`")
    print(f"Unbiased:\n`{tokenizer.decode(unbiased_tokenized_responses[i])}`")
    print()
```

```python
layer = "L00.pkl"
with open(dir / "biased_context" / "acts_q+instr" / layer, "rb") as f:
    biased_acts_qinstr = pickle.load(f)
with open(dir / "unbiased_context" / "acts_q+instr" / layer, "rb") as f:
    unbiased_acts_qinstr = pickle.load(f)
with open(dir / "biased_context" / "acts_resp" / layer, "rb") as f:
    biased_acts_resp = pickle.load(f)
with open(dir / "unbiased_context" / "acts_resp" / layer, "rb") as f:
    unbiased_acts_resp = pickle.load(f)

for i in range(len(tokenized_qs)):
    tokenized_q = tokenized_qs[i].tokenized_question
    biased_tok_resp = biased_tokenized_responses[i]
    unbiased_tok_resp = unbiased_tokenized_responses[i]

    assert (
        len(tokenized_q)
        == biased_acts_qinstr[i].shape[0]
        == unbiased_acts_qinstr[i].shape[0]
    )
    assert len(biased_tok_resp) == biased_acts_resp[i].shape[0]
    assert len(unbiased_tok_resp) == unbiased_acts_resp[i].shape[0]
```
