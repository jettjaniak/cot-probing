```python
from cot_probing.utils import load_task_data
from pprint import pprint
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("google/gemma-2-2b")

baseline_few_shot_prompt, all_a_few_shot_prompt, val_data = load_task_data("snarks")
encoded_baseline_few_shot_prompt = tokenizer.encode(baseline_few_shot_prompt)
print(f"Baseline {len(encoded_baseline_few_shot_prompt)} tokens:")
print(baseline_few_shot_prompt)
```

```python
encoded_all_a_few_shot_prompt = tokenizer.encode(all_a_few_shot_prompt)
print(f"All A {len(encoded_all_a_few_shot_prompt)} tokens:")
print(all_a_few_shot_prompt)
```
